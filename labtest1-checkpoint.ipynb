```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
np.random.seed(0)
traffic_data = np.sin(np.linspace(0,24,1440)) + np.random.random(1440)*10
print(traffic_data)
```

```python
smoothed_data = signal.savgol_filter(traffic_data,polyorder=3,window_length=51)
```

```python
print(smoothed_data)
```

```python
def avg_vech(smoothed_data):
    hour_avg=[]
    
    for i in range(0,1440,60):
        sum=0
        for j in range(0,60,1):
            sum+=smoothed_data[i+j]
        hour_avg.append(sum/60)
    return hour_avg

hour_avg=avg_vech(smoothed_data)
#     return np.mean(smoothed_data,axis=0)
print(avg_vech(smoothed_data))
```

```python
plt.plot(traffic_data,label='noisy data',color='green')
plt.xlabel('Number of minutes')
plt.title('LAB TEST 1')
plt.plot(smoothed_data,color='red')
plt.legend()
plt.show()
```
